```python
#importing dependencies
import matplotlib.pyplot as plt
import numpy as np
```

```python
#data set for the plot
temp = [14.2, 16.4, 11.9, 15.2, 18.5, 22.1, 19.4, 25.1, 23.4, 18.1, 22.6, 17.2]
sales = [215, 325, 185, 332, 406, 522, 412, 614, 544, 421, 445, 408]
```

```python
#ice cream scoop price data
scoop_price = [10, 18, 29, 45, 79, 46, 28, 38, 89, 89, 62, 69]
```

```python
# plot the scatterplot using red dots, and size based off the scoop price
plt.scatter(temp, sales, facecolor="red", s=scoop_price)
```

```python
# limits the x axis to 10-26
plt.xlim(10, 26)
```

```python
# limits the y axis from 100 to 600
plt.ylim(100, 600)
```

```python
# adding the title and axis labels
plt.title("Ice Cream Sales")
plt.xlabel("Temperature Outside")
plt.ylabel("Sales (in scoops)")
```
