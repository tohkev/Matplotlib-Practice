```python
# importing dependencies
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# reading csv data
data = "../Resources/avg_rain_state.csv"
avg_rain_data = pd.read_csv(data)
avg_rain_data.head()
```

```python
#saving specific csv data to variables
states = avg_rain_data["State"]
rainfall = avg_rain_data["Inches"]
```

```python
# creating x axis
x_axis = np.arange(0, len(states), 1)
```

```python
# plotting data into a bar graph
plt.figure(figsize=(20, 3))
plt.bar(x_axis, rainfall, alpha=0.5, align="edge")
```

```python
# adding and adjusting tick labels so that they are spaced and even
tick_loc = [num+0.4 for num in x_axis]
plt.xticks(tick_loc, states, rotation="vertical")
```

```python
# limiting x axis values shown
plt.xlim(-0.25, len(x_axis))
```

```python
# limiting y axis values shown
plt.ylim(0, max(rainfall)+10)
```

```python
#adding labels and title to the chart
plt.xlabel("States in the U.S.")
plt.ylabel("Average Rainfall (in Inches)")
plt.title("Average Rainfall per Year in the U.S.")

#adjusts the view so that the table and it labels are able to be seen
plt.tight_layout()
```
