# New Jersey Weather

```python
# importing dependencies
import numpy as np
import matplotlib.pyplot as plt
```

```python
# average temperature per month in fahrenheit
f_temp = [39, 42, 51, 62, 72, 82, 86, 84, 77, 65, 55, 44]
```

```python
# converting fahrenheit temperatures to celsius
c_temp = [((num-32)*0.56) for num in f_temp]
```

```python
# creating x axis based on the length of the numbers in the list
x_axis = np.arange(0, len(f_temp), 1)
```

```python
# plotting fahrenheit values
plt.plot(x_axis, f_temp)
plt.xlabel("Month")
plt.ylabel("Degrees in Fahrenheit")

plt.show()
```

```python
# plotting celsius values
plt.plot(x_axis, c_temp)
plt.xlabel("Month")
plt.ylabel("Degree in Celsius")

plt.show()
```

```python
# plotting both celsius and fahrenheit values
plt.plot(x_axis, f_temp)
plt.plot(x_axis, c_temp)
plt.xlabel("Month")

plt.show()
```
