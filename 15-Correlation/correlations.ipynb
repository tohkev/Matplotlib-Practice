```python
# importing dependencies
import pandas as pd
import sklearn.datasets as dta
import scipy.stats as st
import matplotlib.pyplot as plt
%matplotlib notebook
```

```python
# reading the wine recognition data set from sklearn and loading into Pandas
data = dta.load_wine()
wine_data = pd.DataFrame(data.data,columns=data.feature_names)
wine_data.head()
```

```python
# Plotting malic_acid versus flavanoids on a scatterplot
malic_acid = wine_data["malic_acid"]
flavanoids = wine_data["flavanoids"]
plt.scatter(malic_acid, flavanoids)
plt.xlabel("Malic Acid Content")
plt.ylabel("Flavanoid Content")
```

```python
# Calculating the correlation coefficient between malic_acid and flavanoids
correlation = st.pearsonr(malic_acid, flavanoids)
print(f"The correlation between Malic Acid and Flavanoids is {round(correlation[0],2)}")
```

```python
# Plotting alcohol versus colour_intensity on a scatterplot
alcohol = wine_data["alcohol"]
color_intensity = wine_data["color_intensity"]
plt.scatter(alcohol, color_intensity)
plt.xlabel("Alcohol Content (%)")
plt.ylabel("Color Intensity")
```

```python
# Calculating the correlation coefficient between alcohol and color_intensity
correlation = st.pearsonr(alcohol, color_intensity)
print(f"The correlation between the alcohol content and color intensity is {round(correlation[0],2)}")
```

```python
# generating the correlation matrix
wine_correlation = wine_data.corr()
wine_correlation.unstack().sort_values()
```
