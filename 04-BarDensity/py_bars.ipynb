```python
import math
import matplotlib.pyplot as plt
import numpy as np
```

```python
cities = ["New Orleans", "Milwaukee", "Omaha", "Pittsburgh", "Toledo"]
bars_in_cities = [8.6, 8.5, 8.3, 7.9, 7.2]
x_axis = np.arange(len(cities))
```

```python
# creating bar chart based upon the above data
plt.bar(x_axis, bars_in_cities, color="b", align="center", alpha=1)
```

```python
# creates the ticks for our bar chart's x axis
tick_locations = [value for value in x_axis]
plt.xticks(tick_locations, cities)
```

```python
# sets the limits of the x axis
plt.xlim(-0.75, len(x_axis)-0.25)
```

```python
# sets the limits of the y axis
plt.ylim(0, math.ceil(max(bars_in_cities)))
```

```python
# gives the chart a title, x label, and y label
plt.title("Density of Bars in Cities")
plt.xlabel("Observed Cities")
plt.ylabel("Bars per 10,000 Households")
```
