```python
# importing dependencies
import numpy as np
import matplotlib.pyplot as plt
```

```python
#average fahrenheit values
avg_f = [39,42,51,62,72,82,86,84,77,65,55,44]
```

```python
#average celsius values from list comprehension
avg_c = [(x-32)*0.56 for x in avg_f]
```

```python
# creating x axis based off of the length of the list
x_axis = np.arange(0, len(avg_f), 1)
```

```python
plt.plot(x_axis, avg_f, marker="x", color="blue", label="Fahrenheit")
plt.plot(x_axis, avg_c, marker="o", color="red", label="Celsius")
plt.legend(loc="upper right")
plt.xlabel("Month")
plt.ylabel("Degrees")

plt.show()
```
