```python
# Import Dependencies
import numpy as np
import matplotlib.pyplot as plt
```

```python
# DATA SET 1
gyms = ["Crunch", "Planet Fitness", "NY Sports Club", "Rickie's Gym"]
members = [49, 92, 84, 53]
```

```python
x_axis = np.arange(0, len(gyms), 1)
tick_loc = [num for num in x_axis]
```

```python
plt.figure(figsize=(10, 5))
plt.bar(x_axis, members, alpha=0.5)

plt.xticks(tick_loc, gyms)
plt.xlabel("Gym")
plt.ylabel("Total Members")
plt.title("Membership in Local Gyms")
plt.xlim(-0.75, len(gyms)-0.25)
plt.ylim(0, max(members)+10)

plt.show()
```

```python
# DATA SET 2
x_lim = 2 * np.pi
x_axis = np.arange(0, x_lim, 0.1)
sin = np.sin(x_axis)
```

```python
plt.plot(x_axis, sin, marker="^", color="g")

plt.title("Sin from 0 to 2 pi")
plt.xlabel("Real Numbers")
plt.ylabel("sin(x)")
plt.hlines(0, 0, 10, alpha=0.25)
plt.grid()

plt.xlim(0, x_lim)
plt.ylim(-1.25, 1.25)

plt.show()
```

```python
# DATA SET 3
gyms = ["Crunch", "Planet Fitness", "NY Sports Club", "Rickie's Gym"]
members = [49, 92, 84, 53]
x_axis = np.arange(0, len(gyms))
colors = ["yellowgreen", "red", "lightcoral", "lightskyblue"]
explode = (0, 0.05, 0, 0)
```

```python
plt.pie(members, explode=explode, labels=gyms, autopct= "%1.1f%%", shadow=True, startangle=140)
plt.axis("equal")

plt.title("Percentage of Community in Local Gyms")

plt.show()
```

```python
# DATA SET 4
x_axis = np.arange(0, 10, 0.1)
times = []
for x in x_axis:
    times.append(x * x + np.random.randint(0, np.ceil(max(x_axis))))
```

```python
plt.scatter(x_axis, times, facecolor="red", edgecolor="black", alpha=0.5)

plt.xlim(-0.75, max(x_axis)+0.25)
plt.ylim(0, max(times)+5)

plt.xlabel("Length of X Axis")
plt.ylabel("Time")
plt.title("Running Time for Sample Outputs")
plt.show()
```
