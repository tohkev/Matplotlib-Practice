```python
# importing dependencies
import matplotlib.pyplot as plt
import numpy as np
```

```python
# average speed of coasters recorded every 10 seconds
anger_drop = [9, 8, 90, 85, 80, 70, 70, 65, 55, 60, 70, 65, 50]
rail_gun = [75, 70, 60, 65, 60, 45, 55, 50, 40, 40, 35, 35, 30]
```

```python
# x_axis seconds
x_axis = np.arange(0, len(danger_drop)*10, 10)
```

```python
plt.plot(x_axis, danger_drop, color="blue", label="Danger Drop")
plt.plot(x_axis, rail_gun, color="red", label="Rail Gun")

#focusing on main dataset
plt.xlim(0, 120)
plt.ylim(0, 100)

# labeling title, x and y axis, and showing grid lines
plt.xlabel("Coaster Runtime")
plt.ylabel("MPH")
plt.title("Coaster Speed Over Time")
plt.grid()
plt.legend(loc="best")

plt.show()
```
